## Задание 1

Найти обратную матрицу методом присоединенной матрицы

$A = 
\begin{pmatrix*}[r]
    5 & \; 3  & \; 1 \\
    1  & \; -3 & \; -2 \\
    -5  & \; 2  & \; 1
\end{pmatrix*}
$

### Решение

Определитель исходной матрицы:

$det(A) = 5 \cdot (-3) \cdot 1 + 3 \cdot (-2) \cdot (-5) + 1 \cdot 2 \cdot 1 - 1 \cdot (-3) \cdot (-5) - 2 \cdot (-2) \cdot 5 - 1 \cdot 3 \cdot 1 = -15 + 30 + 2 - 15 + 20 - 3 = 19$

Матрица $A$ является невырожденной, соответственно, можем найти обратную матрицу

Далее вычислим алгебраические дополнения ко всем элементам матрицы $A$

$A_{11} = (-1)^{1+1} \cdot
\begin{vmatrix*}[r]
    -3 & \; -2 \\
    2  & \; 1 \\
\end{vmatrix*} = 1; \quad$
$A_{12} = (-1)^{1+2} \cdot
\begin{vmatrix*}[r]
    1 & \; -2 \\
    -5  & \; 1 \\
\end{vmatrix*} = 9; \\$

$A_{13} = (-1)^{1+3} \cdot
\begin{vmatrix*}[r]
    1 & \; -3 \\
    -5  & \; 2 \\
\end{vmatrix*} = -13 \quad$
$A_{21} = (-1)^{2+1} \cdot
\begin{vmatrix*}[r]
    3 & \; 1 \\
    2  & \; 1 \\
\end{vmatrix*} = -1; \\$

$A_{22} = (-1)^{2+2} \cdot
\begin{vmatrix*}[r]
    5 & \; 1 \\
    -5  & \; 1 \\
\end{vmatrix*} = 10; \quad$
$A_{23} = (-1)^{2+3} \cdot
\begin{vmatrix*}[r]
    5 & \; 3 \\
    -5  & \; 2 \\
\end{vmatrix*} = -25 \\$

$A_{31} = (-1)^{3+1} \cdot
\begin{vmatrix*}[r]
    3 & \; 1 \\
    -3  & \; -2 \\
\end{vmatrix*} = -3; \quad$
$A_{32} = (-1)^{3+2} \cdot
\begin{vmatrix*}[r]
    5 & \; 1 \\
    1  & \; -2 \\
\end{vmatrix*} = 11; \\$

$A_{33} = (-1)^{3+3} \cdot
\begin{vmatrix*}[r]
    5 & \; 3 \\
    1  & \; -3 \\
\end{vmatrix*} = -18$

Запишем матрицу алгебраических дополнений и транспонируем ее

$A^* =
\begin{pmatrix*}[r]
    1 & \; 9  & \; -13 \\
    -1  & \; 10 & \; -25 \\
    -3  & \; 11  & \; -18
\end{pmatrix*} \implies A^{*T} =
\begin{pmatrix*}[r]
    1 & \; -1  & \; -3 \\
    9  & \; 10 & \; 11 \\
    -13  & \; -25  & \; -18
\end{pmatrix*}$

Найдем обратную матрицу

$A^{-1} = \frac{1}{det(A)} A^{*T} = \frac{1}{19}
\begin{pmatrix*}[r]
    1 & \; -1  & \; -3 \\
    9  & \; 10 & \; 11 \\
    -13  & \; -25  & \; -18
\end{pmatrix*} = 
\begin{pmatrix*}[r]
    \frac{1}{19} & \; -\frac{1}{19}  & \; -\frac{3}{19} \\
    \\ \frac{9}{19}  & \; \frac{10}{19} & \; \frac{11}{19} \\
    \\ -\frac{13}{19}  & \; -\frac{25}{19}  & \; -\frac{18}{19}
\end{pmatrix*}$

## Задание 2

Решить матричное уравнение

$\begin{pmatrix*}[r]
    1 & \; -2  & \; 3 \\
    2  & \; 3 & \; -1 \\
    0  & \; -2  & \; 1
\end{pmatrix*} \cdot X \cdot
\begin{pmatrix*}[r]
    1 & \; 2  & \; 3 \\
    4  & \; 5 & \; 6 \\
    7  & \; 8  & \; 0
\end{pmatrix*} =
\begin{pmatrix*}[r]
    1 & \; 2  & \; 3 \\
    4  & \; 5 & \; 6 \\
    7  & \; 8  & \; 0
\end{pmatrix*}$

### Решение

В общем виде, уравнение выглядит следующим образом:

$A \cdot X \cdot B = B$

Для начала, определим, являются ли матрицы в левой части уравнения невырожденными

$det(A) = 1 \cdot 3 \cdot 1 + (-2) \cdot (-1) \cdot 0 + 2 \cdot (-2) \cdot 3 - 3 \cdot 3 \cdot 0 - 1 \cdot (-1) \cdot (-2) - 2 \cdot (-2) \cdot 1 = 3 - 12 - 2 + 4 = -7$

$det(B) = 1 \cdot 5 \cdot 0 + 2 \cdot 6 \cdot 7 + 4 \cdot 8 \cdot 3 - 3 \cdot 5 \cdot 7 - 2 \cdot 4 \cdot 0 - 6 \cdot 8 \cdot 1 = 84 + 96 - 105 - 48 = 27$

Обе матрицы являются невырожденными, соответственно, решение уравнения записывается следующим образом:

$X = B \cdot B^{-1} \cdot A^{-1} = A^{-1}$

Вычислим алгебраические дополнения ко всем элементам матрицы $A$

$A_{11} = (-1)^{1+1} \cdot
\begin{vmatrix*}[r]
    3 & \; -1 \\
    -2  & \; 1 \\
\end{vmatrix*} = 1; \quad$
$A_{12} = (-1)^{1+2} \cdot
\begin{vmatrix*}[r]
    2 & \; -1 \\
    0  & \; 1 \\
\end{vmatrix*} = -2; \\$

$A_{13} = (-1)^{1+3} \cdot
\begin{vmatrix*}[r]
    2 & \; 3 \\
    0  & \; -2 \\
\end{vmatrix*} = -4 \quad$
$A_{21} = (-1)^{2+1} \cdot
\begin{vmatrix*}[r]
    -2 & \; 3 \\
    -2  & \; 1 \\
\end{vmatrix*} = -4; \\$

$A_{22} = (-1)^{2+2} \cdot
\begin{vmatrix*}[r]
    1 & \; 3 \\
    0  & \; 1 \\
\end{vmatrix*} = 1; \quad$
$A_{23} = (-1)^{2+3} \cdot
\begin{vmatrix*}[r]
    1 & \; -2 \\
    0  & \; -2 \\
\end{vmatrix*} = 2 \\$

$A_{31} = (-1)^{3+1} \cdot
\begin{vmatrix*}[r]
    -2 & \; 3 \\
    3  & \; -1 \\
\end{vmatrix*} = -7; \quad$
$A_{32} = (-1)^{3+2} \cdot
\begin{vmatrix*}[r]
    1 & \; 3 \\
    2  & \; -1 \\
\end{vmatrix*} = 7; \\$

$A_{33} = (-1)^{3+3} \cdot
\begin{vmatrix*}[r]
    1 & \; -2 \\
    2  & \; 3 \\
\end{vmatrix*} = 7$

Запишем матрицу алгебраических дополнений и транспонируем ее

$A^* =
\begin{pmatrix*}[r]
    1 & \; -2  & \; -4 \\
    -4  & \; 1 & \; 2 \\
    -7  & \; 7  & \; 7
\end{pmatrix*} \implies A^{*T} =
\begin{pmatrix*}[r]
    1 & \; -4  & \; -7 \\
    -2  & \; 1 & \; 7 \\
    -4  & \; 2  & \; 7
\end{pmatrix*}$

Тогда

$X = A^{-1} = \frac{1}{det(A)} A^{*T} = \frac{1}{-7}
\begin{pmatrix*}[r]
    1 & \; -4  & \; -7 \\
    -2  & \; 1 & \; 7 \\
    -4  & \; 2  & \; 7
\end{pmatrix*} = 
\begin{pmatrix*}[r]
    -\frac{1}{7} & \; \frac{4}{7}  & \; 1 \\
    \\ \frac{2}{7}  & \; -\frac{1}{7} & \; -1 \\
    \\ \frac{4}{7}  & \; -\frac{2}{7}  & \; -1
\end{pmatrix*}$

## Задание 3

Найти ранг матрицы при различных значениях параметра

$
\begin{pmatrix}
    3 & \; 1  & \; 1 & \; 4 \\
    \lambda & \; 4  & \; 10 & \; 1 \\
    1 & \; 7  & \; 17 & \; 3 \\
    2 & \; 2  & \; 4 & \; 3
\end{pmatrix}
$

### Решение

Т.к. в матрице присутствуют ненулевые элементы, то $r \geq 1$

Найдем ненулевой минор второго порядка. Например

$M_2 = 
\begin{vmatrix*}[r]
    1 & \; 1 \\
    4  & \; 10 \\
\end{vmatrix*} = 6$, соответственно, $r \geq 2$

Для минора $M_2$ найдем окаймляющие миноры третьего порядка

$M_3 = 
\begin{vmatrix*}[r]
    1 & \; 1 & \; 4 \\
    4  & \; 10 & \; 1 \\
    7  & \; 17 & \; 3 \\
\end{vmatrix*} = 30 + 7 + 272 - 280 - 17 - 12 = 0$

$M_3 = 
\begin{vmatrix*}[r]
    1 & \; 1 & \; 4 \\
    4  & \; 10 & \; 1 \\
    2  & \; 4 & \; 3 \\
\end{vmatrix*} = 30 + 2 + 64 - 80 - 4 - 12 = 0$

$M_3 = 
\begin{vmatrix*}[r]
    3 & \; 1 & \; 1 \\
    \lambda  & \; 4 & \; 10 \\
    1  & \; 7 & \; 17 \\
\end{vmatrix*} = 204 + 10 + 7 \lambda - 4 - 210 - 17 \lambda = -10 \lambda$

$M_3 = 
\begin{vmatrix*}[r]
    3 & \; 1 & \; 1 \\
    \lambda  & \; 4 & \; 10 \\
    2  & \; 2 & \; 4 \\
\end{vmatrix*} = 48 + 20 + 2 \lambda - 8 - 60 - 4 \lambda = -2 \lambda$

Соответственно, при $\lambda = 0, \; r = 2$, т.к. отсутствуют ненулевые окаймляющие миноры третьего порядка

Теперь проверим, при каком параметре $\lambda$, ранг исходной матрицы будет равен 4. Посчитаем определитель, разложив матрицу по элементам первого столбца

$
\begin{vmatrix}
    3 & \; 1  & \; 1 & \; 4 \\
    \lambda & \; 4  & \; 10 & \; 1 \\
    1 & \; 7  & \; 17 & \; 3 \\
    2 & \; 2  & \; 4 & \; 3
\end{vmatrix} = 3
\begin{vmatrix*}[r]
    4 & \; 10 & \; 1 \\
    7  & \; 17 & \; 3 \\
    2  & \; 4 & \; 3 \\
\end{vmatrix*} - \lambda
\begin{vmatrix*}[r]
    1 & \; 1 & \; 4 \\
    7  & \; 17 & \; 3 \\
    2  & \; 4 & \; 3 \\
\end{vmatrix*} + 1
\begin{vmatrix*}[r]
    1 & \; 1 & \; 4 \\
    4  & \; 10 & \; 1 \\
    2  & \; 4 & \; 3 \\
\end{vmatrix*} - 2
\begin{vmatrix*}[r]
    1 & \; 1 & \; 4 \\
    4  & \; 10 & \; 1 \\
    7  & \; 17 & \; 3 \\
\end{vmatrix*} = 0$

Определитель равен 0 при любом значении $\lambda$, т.к. все алгебраические дополнения равны 0. Соответственно, $r \neq 4$.

Таким образом, ответ: $r = 
\begin{cases}
    2  & \quad \text{при } \lambda = 0 \\
    3  & \quad \text{при } \lambda \neq 0
\end{cases}$

## Задание 4

Исследовать систему и найти решение в зависимости от значения параметра

$
\begin{cases}
    x_1 + 4x_2 + 2x_3 = -1 \\
    2x_1 + 3x_2 - x_3 = 3 \\
    x_1 - x_2 - 3x_3 = 4 \\
    x_1 - 6x_2 - \lambda x_3 = 9 \\
\end{cases}
$

### Решение

Найдем ранг расширенной матрицы

$r(A|b) = 
\begin{pmatrix*}[r]
    1 & \; 4  & \; 2 & \; \big | & \; -1 \\
    2 & \; 3  & \; -1 & \; \big | & \; 3 \\
    1 & \; -1  & \; -3 & \; \big | & \; 4 \\
    1 & \; -6  & \; - \lambda & \; \big | & \; 9
\end{pmatrix*} \quad \Bigg| \quad -2(1);\quad (1) += (2)$

$\begin{pmatrix*}[r]
    0 & \; -5  & \; -5 & \; \big | & \; 5 \\
    2 & \; 3  & \; -1 & \; \big | & \; 3 \\
    1 & \; -1  & \; -3 & \; \big | & \; 4 \\
    1 & \; -6  & \; - \lambda & \; \big | & \; 9
\end{pmatrix*} \quad \Bigg| \quad -1(2);\quad (2) += 2(3)$

$\begin{pmatrix*}[r]
    0 & \; -5  & \; -5 & \; \big | & \; 5 \\
    0 & \; -5  & \; -5 & \; \big | & \; 5 \\
    1 & \; -1  & \; -3 & \; \big | & \; 4 \\
    1 & \; -6  & \; - \lambda & \; \big | & \; 9
\end{pmatrix*} \quad \Bigg| \quad (1) += (2)$

$\begin{pmatrix*}[r]
    0 & \; 0  & \; 0 & \; \big | & \; 0 \\
    0 & \; -5  & \; -5 & \; \big | & \; 5 \\
    1 & \; -1  & \; -3 & \; \big | & \; 4 \\
    1 & \; -6  & \; - \lambda & \; \big | & \; 9
\end{pmatrix*} \quad \Bigg| \quad -1(3);\quad (3) += (4)$

$\begin{pmatrix*}[r]
    0 & \; 0  & \; 0 & \; \big | & \; 0 \\
    0 & \; -5  & \; -5 & \; \big | & \; 5 \\
    0 & \; -5  & \; 3 - \lambda & \; \big | & \; 5 \\
    1 & \; -6  & \; - \lambda & \; \big | & \; 9
\end{pmatrix*}$

Из получившейся расширенной матрицы видим, что

$r(A|b) = r(A) = 
\begin{cases}
    2  & \quad \text{при } \lambda = 8 \\
    3  & \quad \text{при } \lambda \neq 8
\end{cases}$

Таким образом, система всегда является совместной.

При $\lambda \neq 8 \rightarrow r(A|b) = r(A) = n$, (где n - число неизвестных) - система имеет единственное решение.

При $\lambda = 8 \rightarrow r(A|b) = r(A) < n$ - система имеет общее решение.

Найдем общее решение при $\lambda = 8$

$\begin{pmatrix*}[r]
    0 & \; 0  & \; 0 & \; \big | & \; 0 \\
    0 & \; 0  & \; 0 & \; \big | & \; 0 \\
    0 & \; -5  & \; -5 & \; \big | & \; 5 \\
    1 & \; -6  & \; -8 & \; \big | & \; 9
\end{pmatrix*}$

Выберем базисный минор по $x_1, \; x_2$

$\begin{vmatrix}
    0 & \; -5 \\
    1 & \; -6
\end{vmatrix}$

Таким образом

$\begin{cases}
    -5x_2 = 5 + 5x_3 \\
    x_1 - 6x_2 = 9 + 8x_3
\end{cases}$

$\begin{cases}
    x_2 = -1 - x_3 \\
    x_1 = 3 + 2x_3
\end{cases}$

Общее решение: $
\begin{pmatrix}
    3 + 2x_3 \\
    -1 - x_3 \\
    x_3
\end{pmatrix}$

Найдем решение при $\lambda \neq 8$

$\begin{pmatrix*}[r]
    0 & \; 0  & \; 0 & \; \big | & \; 0 \\
    0 & \; -5  & \; -5 & \; \big | & \; 5 \\
    0 & \; -5  & \; 3 - \lambda & \; \big | & \; 5 \\
    1 & \; -6  & \; - \lambda & \; \big | & \; 9
\end{pmatrix*} \quad \Bigg| \quad (2) += (3)$

$\begin{pmatrix*}[r]
    0 & \; 0  & \; 0 & \; \big | & \; 0 \\
    0 & \; 0  & \; \lambda - 8 & \; \big | & \; 0 \\
    0 & \; -5  & \; 3 - \lambda & \; \big | & \; 5 \\
    1 & \; -6  & \; - \lambda & \; \big | & \; 9
\end{pmatrix*}$

Базисный минор по $x_1, \; x_2, \; x_3$

$\begin{vmatrix}
    0 & \; 0 & \; \lambda - 8 \\
    0 & \; -5 & \; 3 - \lambda \\
    1 & \; -6 & \; - \lambda
\end{vmatrix}$

Получаем

$
\begin{cases}
    (\lambda - 8)x_3 = 0 \\
    -5x_2 + (3 - \lambda)x_3 = 5 \\
    x_1 - 6x_2 - \lambda x_3 = 9
\end{cases}$

Откуда $x_1 = 3, \; x_2 = -1, \; x_3 = 0$

**Ответ:** При $\lambda \neq 8$ система совместна и имеет единственное решение $x_1 = 3, \; x_2 = -1, \; x_3 = 0$. При $\lambda = 8$ система совместна и имеет общее решение $
\begin{pmatrix}
    3 + 2x_3 \\
    -1 - x_3 \\
    x_3
\end{pmatrix}$

## Задание 5

$
\begin{pmatrix*}[r]
    1 & \; -3 & \; 4 \\

    6 & \; -7 & \; 7
\end{pmatrix*}$

### Решение

Запишем характеристическое уравнение матрицы

$
\begin{vmatrix}
    1 - \lambda & \; -3 & \; 4 \\
    4 & \; -7 - \lambda & \; 8 \\
    6 & \; -7 & \; 7 - \lambda
\end{vmatrix} = 0$

$(1 - \lambda) \big( (7 - \lambda)(-7 - \lambda) + 56 \big ) - 4 \big(-3(7 - \lambda) + 28 \big) + 6 \big(-24-4(-7 - \lambda) \big) = 0$

$(1 - \lambda)(-49 - 7 \lambda + 7 \lambda + \lambda^2 + 56) - 4(-21 + 3 \lambda + 28) + 6(-24 + 28 + 4 \lambda) = 0$

$\lambda^2 + 7 - \lambda^3 - 7 \lambda - 12 \lambda - 28 + 24 \lambda + 24 = 0$

$- \lambda^3 + \lambda^2 + 5 \lambda + 3 = 0$

$(\lambda + 1)(\lambda + 1)(\lambda - 3) = 0$

Корнями данного уравнения и собственными значениями матрицы являются $\lambda = -1$ и $\lambda = 3$

Найдем вид собственных векторов для собственного значения $\lambda = -1$

$
\begin{pmatrix*}[r]
    2 & \; -3 & \; 4 \\
    4 & \; -6 & \; 8 \\

\end{pmatrix*} \quad \Bigg| \quad 2(1); \quad (1) -= (2)$

$

\end{pmatrix*} \quad \Bigg| \quad -1(3row)$

$
\begin{pmatrix*}[r]
    0 & \; 0 & \; 0 \\

    6 & \; -7 & \; -8
\end{pmatrix*}$

$\begin{vmatrix}

\end{vmatrix}$

Получаем

Находим общее решение $x_1 = x_3, \; x_2 = 2x_3, \; x_3$
