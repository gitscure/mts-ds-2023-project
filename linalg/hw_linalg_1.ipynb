## Задание 1

Зная, что $|\vec{a}| = 2$, $|\vec{b}| = 5$ и $\widehat{(a,b)} = \frac{2\pi}{3}$, определить, при каком значении коэффициента $\alpha$ векторы $\vec{p} = \alpha\vec{a} + 17\vec{b}$ и $\vec{q} = 3\vec{a} - \vec{b}$ окажутся перпендикулярными.

### Решение

Векторы перпендикулярны, если их скалярное произведение равно 0:

$(\vec{a}, \vec{b}) = 0$

$(\alpha\vec{a} + 17\vec{b})(3\vec{a} - \vec{b}) = 0$

$3\alpha\vec{a}^2 - \alpha\vec{a}\vec{b} + 51\vec{a}\vec{b} - 17\vec{b}^2 = 0$

$3\alpha|\vec{a}|^2 - \alpha|\vec{a}||\vec{b}|cos(\frac{2\pi}{3}) + 51|\vec{a}||\vec{b}|cos(\frac{2\pi}{3}) - 17|\vec{b}|^2 = 0$

$12\alpha - 10\alpha(-0,5) + 510(-0,5) - 425 = 0$

$17\alpha - 680 = 0$

$\alpha = 40$

**Ответ:** $\alpha = 40$

## Задание 2

Пусть $G$ - множество всех вектор-столбцов линейного пространства $R^n$ с положительными элементами, то есть

$G = \{x:x = 
 \begin{pmatrix}
  x_1 \\
  x_2 \\
  \cdots \\
  x_n 
 \end{pmatrix}
 \in R^n, x_i > 0, i = 1,2,\dotso,n\}$

 Проверить, является ли линейным пространством множество $G$, если операции сложения векторов и умножения вектора на число определяются следующим образом

 $x + y = 
 \begin{pmatrix}
  x_1 \cdot y_1 \\
  x_2 \cdot y_2 \\
  \cdots \\
  x_n \cdot y_n 
 \end{pmatrix}$
 и $\alpha x = 
 \begin{pmatrix}
  x_1^{\alpha} \\
  x_2^{\alpha} \\
  \cdots \\
  x_n^{\alpha}
 \end{pmatrix}$

### Решение

0. Для начала проверим корректность определения операций сложения векторов и умножения вектора на число. Так, произведение вещественных чисел $x_n \cdot y_n$ всегда положительно, а возведение вещественного числа в степень $x_n^{\alpha}$ также является положительным вещественным числом. Здесь и далее принимаем, что $\alpha \in R$.

Проверим аксиомы заданного линейного пространства

1. Коммутативность сложения

$x + y = 
 \begin{pmatrix}
  x_1 \cdot y_1 \\
  x_2 \cdot y_2 \\
  \cdots \\
  x_n \cdot y_n 
 \end{pmatrix} = y + x$ для $\forall x, y \in G$.

2. Ассоциативность сложения.

$(x + y) + z = 
\begin{pmatrix}
  (x_1 \cdot y_1) \cdot z_1 \\
  (x_2 \cdot y_2) \cdot z_2 \\
  \cdots \\
  (x_n \cdot y_n) \cdot z_n 
 \end{pmatrix} = 
 \begin{pmatrix}
  x_1 \cdot (y_1 \cdot z_1) \\
  x_2 \cdot (y_2 \cdot z_2) \\
  \cdots \\
  x_n \cdot (y_n \cdot z_n) 
 \end{pmatrix} = x + (y + z)$ для $\forall x, y, z \in G$

3. Существование нейтрального элемента относительно сложения

$x + 0 = 
\begin{pmatrix}
  x_1 \\
  x_2 \\
  \cdots \\
  x_n 
 \end{pmatrix} +
 \begin{pmatrix}
  1 \\
  1 \\
  \cdots \\
  1 
 \end{pmatrix} = 
 \begin{pmatrix}
  x_1 \cdot 1 \\
  x_2 \cdot 1 \\
  \cdots \\
  x_n \cdot 1
 \end{pmatrix} = x$ для $\forall x \in G$

4. Существование обратного элемента относительно сложения

$x + x^{'} = 
\begin{pmatrix}
  x_1 \\
  x_2 \\
  \cdots \\
  x_n 
 \end{pmatrix} +
 \begin{pmatrix}
  x_1^{-1} \\
  x_2^{-1} \\
  \cdots \\
  x_n^{-1} 
 \end{pmatrix} = 
 \begin{pmatrix}
  x_1 \cdot x_1^{-1} \\
  x_2 \cdot x_2^{-1} \\
  \cdots \\
  x_n \cdot x_n^{-1} 
 \end{pmatrix} = 
 \begin{pmatrix}
  1 \\
  1 \\
  \cdots \\
  1 
 \end{pmatrix} = 0$ для $\forall x \in G$

5. Дистрибутивность по сложению векторов

$\alpha(x + y) = \alpha
\begin{pmatrix}
  x_1 \cdot y_1 \\
  x_2 \cdot y_2 \\
  \cdots \\
  x_n \cdot y_n 
\end{pmatrix} = 
\begin{pmatrix}
  (x_1 \cdot y_1)^{\alpha} \\
  (x_2 \cdot y_2)^{\alpha} \\
  \cdots \\
  (x_n \cdot y_n)^{\alpha} 
\end{pmatrix} = 
\begin{pmatrix}
  x_1^{\alpha} \cdot y_1^{\alpha} \\
  x_2^{\alpha} \cdot y_2^{\alpha} \\
  \cdots \\
  x_n^{\alpha} \cdot y_n^{\alpha} 
\end{pmatrix} = 
\begin{pmatrix}
  x_1^{\alpha} \\
  x_2^{\alpha} \\
  \cdots \\
  x_n^{\alpha} 
\end{pmatrix} + 
\begin{pmatrix}
  y_1^{\alpha} \\
  y_2^{\alpha} \\
  \cdots \\
  y_n^{\alpha} 
\end{pmatrix} = \alpha
\begin{pmatrix}
  x_1 \\
  x_2 \\
  \cdots \\
  x_n 
\end{pmatrix} + \alpha
\begin{pmatrix}
  y_1 \\
  y_2 \\
  \cdots \\
  y_n
\end{pmatrix} = 
\alpha x + \alpha y$

для $\forall x, y \in G$ и $\forall \alpha \in R$

6. Дистрибутивность по сложению скаляров

$(\alpha + \beta)x = 
\begin{pmatrix}
  x_1^{\alpha + \beta} \\
  x_2^{\alpha + \beta} \\
  \cdots \\
  x_n^{\alpha + \beta} 
\end{pmatrix} = 
\begin{pmatrix}
  x_1^{\alpha} \cdot x_1^{\beta} \\
  x_2^{\alpha} \cdot x_2^{\beta} \\
  \cdots \\
  x_n^{\alpha} \cdot x_n^{\beta} 
\end{pmatrix} = 
\begin{pmatrix}
  x_1^{\alpha} \\
  x_2^{\alpha} \\
  \cdots \\
  x_n^{\alpha} 
\end{pmatrix} + 
\begin{pmatrix}
  x_1^{\beta} \\
  x_2^{\beta} \\
  \cdots \\
  x_n^{\beta} 
\end{pmatrix} = \alpha x + \beta x$

для $\forall x \in G$ и $\forall \alpha, \beta \in R$

7. Ассоциативность по умножению скаляров

$(\alpha \beta)x = 
\begin{pmatrix}
  x_1^{\alpha \beta} \\
  x_2^{\alpha \beta} \\
  \cdots \\
  x_n^{\alpha \beta} 
\end{pmatrix} = 
\begin{pmatrix}
  (x_1^{\beta})^{\alpha} \\
  (x_2^{\beta})^{\alpha} \\
  \cdots \\
  (x_n^{\beta})^{\alpha} 
\end{pmatrix} = \alpha
\begin{pmatrix}
  x_1^{\beta} \\
  x_2^{\beta} \\
  \cdots \\
  x_n^{\beta} 
\end{pmatrix} = \alpha(\beta x)$

для $\forall x \in G$ и $\forall \alpha, \beta \in R$

8. Существование нейтрального элемента относительно умножения

$1 \cdot x = 
\begin{pmatrix}
  x_1^{1} \\
  x_2^{1} \\
  \cdots \\
  x_n^{1} 
\end{pmatrix} = 
\begin{pmatrix}
  x_1 \\
  x_2 \\
  \cdots \\
  x_n 
\end{pmatrix} = x$ для $\forall x \in G$

Аксиомы линейного пространства выполняются, соответственно, множество $G$ является линейным пространством.

## Задание 3

Используя определение, доказать, что для любых векторов $x,\; y,\; z$ и чисел $\alpha,\; \beta,\; \gamma$ векторы $\alpha x - \beta y,\; \gamma y - \alpha z,\; \beta z - \gamma x$ линейно зависимы.

### Решение

По определению, вектора являются линейно зависимыми, если существует нетривиальная комбинация векторов $\alpha x - \beta y,\; \gamma y - \alpha z,\; \beta z - \gamma x$ с некоторыми коэффициентами $k_1,\; k_2,\; k_3$ хотя бы один из которых ненулевой, равная нулевому вектору. Иныим словами,

$k_1(\alpha x - \beta y) + k_2(\gamma y - \alpha z) + k_3(\beta z - \gamma x) = 0$.

Выполним некоторые преобразования:

$k_1 \alpha x - k_1 \beta y + k_2 \gamma y - k_2 \alpha z + k_3 \beta z - k_3\gamma x = 0\\$
$(k_1 \alpha - k_3\gamma)x + (k_2 \gamma - k_1 \beta)y + (k_3 \beta - k_2 \alpha)z = 0$


Нас не интересует решение, при котором векторы $x,\; y,\; z$ будут нулевыми, соответственно, коэффициенты при них должны равняться 0. Получаем

$\begin{cases}
  k_1 \alpha - k_3\gamma = 0 \\
  k_2 \gamma - k_1 \beta = 0 \\
  k_3 \beta - k_2 \alpha = 0
\end{cases}$

Таким образом, существуют ненулевые коэффициенты $k_1 = k_3 \frac{\gamma}{\alpha},\; k_2 = k_3 \frac{\beta}{\alpha},\; k_3$, линейная комбинация которых с векторами $\alpha x - \beta y,\; \gamma y - \alpha z,\; \beta z - \gamma x$ будет равна нулевому вектору.

$k_3 \gamma x - k_3 \frac{\gamma \beta y}{\alpha} + k_3 \frac{\gamma \beta y}{\alpha} - k_3 \beta z + k_3 \beta z - k_3 \gamma x = 0\\$
$0 = 0$

Соответственно, для любых векторов $x,\; y,\; z$ и чисел $\alpha,\; \beta,\; \gamma$ векторы $\alpha x - \beta y,\; \gamma y - \alpha z,\; \beta z - \gamma x$ линейно зависимы.

## Задание 5

Найти какой-нибудь базис и размерность линейного пространства $V$, заданного следующим образом: пространство многочленов $p(x) \in P_4$ таких, что $p(1) + p(-1) = 0$

### Решение

Многочлен $p(x)$ можно представить в виде

$p(x) = a_0 + a_1x + a_2x^2 + a_3x^3 + a_4x^4$

Из условия $p(1) + p(-1) = 0$ получаем

$a_0 + a_1 + a_2 + a_3 + a_4 + a_0 - a_1 + a_2 - a_3 + a_4 = 0$

$2a_0 + 2a_2 + 2a_4 = 0 \quad \big| \quad /2$

$a_0 + a_2 + a_4 = 0 \quad \to \quad a_0 = -a_2 - a_4$

Тогда многочлен $p(x)$ представляется в виде:

$p(x) = -a_2 - a_4 + a_1x + a_2x^2 + a_3x^3 + a_4x^4$

$p(x) = a_1x + a_2(x^2 - 1) + a_3x^3 + a_4(x^4 - 1)$

Следовательно, мночлен $p(x)$ представляется комбинацией многочленов $x,\; x^2 - 1,\; x^3,\; x^4 - 1$ с коэффициентами $a_1,\; a_2,\; a_3,\; a_4$

Проверим получившуюся систему на предмет линейной зависимости

$a_1x + a_2(x^2 - 1) + a_3x^3 + a_4(x^4 - 1) = 0$

$a_1(0 + x + 0x^2 + 0x^3 + 0x^4) + a_2(-1 + 0x + x^2 + 0x^3 + 0x^4) + a_3(0 + 0x + 0x^2 + x^3 + 0x^4) + a_4(-1 + 0x + 0x^2 + 0x^3 + x^4) = 0$

Система линейных уравнений относительно коэффициентов $a_1,\; a_2,\; a_3,\; a_4$ выглядит следующим образом 

$
\begin{cases}
  -a_2 - a_4 = 0 \\
  a_1 = 0 \\
  a_2 = 0 \\
  a_3 = 0 \\
  a_4 = 0
\end{cases}
$

Следовательно, равенство

$a_1x + a_2(x^2 - 1) + a_3x^3 + a_4(x^4 - 1) = 0$

достигается при $a_1 = a_2 = a_3 = a_4 = 0$.

А из этого следует, что система многочленов $p_1 = x,\; p_2 = x^2 - 1,\; p_3 = x^3,\; p_4 = x^4 - 1$ является линейно независимой и, соответственно, является базисом пространства $V$ размерностью 4.
